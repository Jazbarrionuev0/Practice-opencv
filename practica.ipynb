```python
img_paisaje = cv2.imread("New_Zealand_Lake.jpg",cv2.IMREAD_COLOR)
# las divide en canales individuales
b, g, r = cv2.split(img_paisaje)

#mostrar cada canal
plt.figure(figsize=[20, 5])
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title('Red cannel');
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title('Green cannel');
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title('Blue cannel');

#juntamos todos los canales 
img_merge = cv2.merge((b,g,r))
plt.subplot(144); plt.imshow(img_merge[:, :,::-1]); plt.title('Merged cannel');
```

```python
#convertir una imafen bgr a rgb
img_convert = cv2.cvtColor(img_paisaje, cv2.COLOR_BGR2RGB)
plt.imshow(img_convert)
```

# MANIPULAR IMAGENES


cambiar pixeles

```python
cubo18_copia = cubo18.copy()

#especifico las rows y columnas donde estan los pixeles a cambiar
cubo18_copia[2,2] = 200
cubo18_copia[2,3] = 200
cubo18_copia[3,2] = 200
cubo18_copia[3,3] = 200

plt.imshow(cubo18_copia, cmap='gray')
print(cubo18_copia)
```

recortar

```python
img_bote = cv2.imread("New_Zealand_Boat.jpg",cv2.IMREAD_COLOR)

img_bote_convierto = cv2.cvtColor(img_bote, cv2.COLOR_BGR2RGB)
plt.imshow(img_bote_convierto)
```

```python
img_bote_recortada = img_bote_convierto[200:400, 300:600]
plt.imshow(img_bote_recortada)
```

```python
img_bote_rerecortada = img_bote_recortada[40:125, 100:170]
plt.imshow(img_bote_rerecortada)
```

dibujar imagen

```python
img_bote_dibujada = cv2.line(img_bote_convierto, (200,400),  (100,500), (0, 255, 255), thickness=5 )

plt.imshow(img_bote_dibujada)
```

```python
img_bote_circulo = cv2.circle(img_bote_convierto, (600, 200), 100, (0, 255, 255), thickness=5 )

plt.imshow(img_bote_circulo)
```

```python
img_bote_rectangulo = cv2.rectangle(img_bote,(400, 350), (500, 250), (0, 255, 255), thickness=5 )

plt.imshow(img_bote_rectangulo)
```
